```python
df.drop(columns=['months_as_customer','policy_bind_date','policy_state','policy_csl','insured_sex','insured_education_level','insured_occupation','insured_hobbies', 
                 'insured_relationship','incident_date','incident_type','collision_type','incident_severity','authorities_contacted',
                 'incident_state','incident_city','incident_location','property_damage','police_report_available',
                 'auto_make','auto_model','_c39','policy_deductable','capital-loss','capital-gains'],inplace=True)
```

```python
df['fraud_reported']=df['fraud_reported'].map({'Y':1,'N':0})
```

```python
corr=df.corr()
corr
```

```python
sns.heatmap(data=corr,annot=True)
plt.show()
```

```python
df.head()
```

```python
x=df.drop('fraud_reported',axis=1)
```

```python
y=df['fraud_reported']
```

```python
plt.figure(figsize = (25, 20))
plotnumber = 1

for col in x.columns:
    if plotnumber <= 24:
        ax = plt.subplot(5, 5, plotnumber)
        sns.distplot(x[col])
        plt.xlabel(col, fontsize = 15)

    plotnumber += 1

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10, 8))
plotnumber = 1

for col in x.columns:
    if plotnumber <= 24:
        ax = plt.subplot(5,5, plotnumber)
        sns.boxplot(x[col])
        plt.xlabel(col, fontsize = 8)

    plotnumber += 1
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train= scaler.fit_transform(x_train)
x_train= pd.DataFrame(X_train, columns =x.columns)
X_test=scaler.fit_transform(x_test)
x_test= pd.DataFrame(X_test, columns =x.columns)
```

```python
#Decision Tree
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
```

```python
y_pred=dtc.predict(x_test)
dtc_train_acc=accuracy_score(y_train,dtc.predict(x_train))
dtc_test_acc=accuracy_score(y_test,y_pred)
```

```python
y_pred
```

```python
print(f"Training accuracy of Decision Tree is : {dtc_train_acc}")
print(f"Test accuracy of Decision Tree is : {dtc_test_acc}")

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
y_pred=dtc.predict([[41,246435,1800.76,0,441499,0,3,1,0,47800,4780,4780,38240,2009]])
```

```python
#RandomForest
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(criterion= 'entropy', max_depth= 10, max_features= 'sqrt', min_samples_leaf= 1, min_samples_split= 3, n_estimators= 140)
rfc.fit(x_train, y_train)
```

```python
y_pred = rfc.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
rand_clf_train_acc = accuracy_score(y_train, rfc.predict(x_train))
rand_clf_test_acc = accuracy_score(y_test, y_pred)
```

```python
print(f"Training accuracy of Random Forest is : {rand_clf_train_acc}")
print(f"Test accuracy of Random Forest is : {rand_clf_test_acc}")

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#KNN
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 30)
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
knn_train_acc = accuracy_score(y_train, knn.predict(x_train))
knn_test_acc = accuracy_score(y_test, y_pred)

print(f"Training accuracy of KNN is : {knn_train_acc}")
print(f"Test accuracy of KNN is : {knn_test_acc}")

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
